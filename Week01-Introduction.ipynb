# Defining Distributed Systems

* A system in which hardware or software components located at networked computers(could be sensors) communicate and coordinate their actions only by message passing.
* A distributed system is a collection of independent computers that appear to the users of the system as a single computer.

Example distributed systems:

* Cluster: A type of parallel or distributed processing system, which consists of a collection of interconnected **stand-alone** computers cooperatively working together as a single, integrated computing resource.
* Cloud: A type of parallel and distributed system consisting of a collection of interconnected and **virtualised computers**(can move from one place to another) that are dynamically provisioned and presented as one or more unified computing resources based on service-level agreements established through negotiation between the service provider and consumers.

# Networks and Distributed Systems

* Networks: A media for interconnecting local and wide area computers and exchange messages based on protocols. Network entities are visible and they are explicitly addressed(IP address).
* Distributed system: existence of multiple autonomous(any computer can disconnect at any time) computers is transparent.

Networks focuses on router, hardware, networking elements and TCP/IP. While the rest of whatever we build on top of it are distributed systems(focus on applications).

# Reasons for Distributed Systems

* Function separation: Existence of computers with different capabilities and purposes
    * Clients and Servers
    * Data collection and data processing
* Inherent distribution
    * Information: They are naturally distributed. Different information is created and maintained by different people. They are available remotely through www.
    * People: They want to work from home and computer can support collaborative work.
* Power imbalance and load variation
    * Distribute computational load among different computers
* Reliability
    * Long term preservation and data backup at different locations. Remember that each computer is autonomous and can fail at any time. When this happens, the request can be transfered to some other working server
* Economies
    * Sharing a printer by many users and reduce the cost of ownership(one computer, one printer, printer server, file server, user server and authentication management)
    * Building a supercomputer out of a network of computers
    
# Consequence of Distributed Systems

Computers in distributed systems may be on separate continents, in the same building, or the same room. DSs have the following consequences(problems need to be solved when choosing DS):

* Concurrency - each system is autonomous and many things are happening in the same time
    * carry out tasks independently
    * tasks coordinate their actions by exchanging messages
* Heterogeneity - Everybody has different kind of computers, different hardware, different software
* No global clock - We can't assume of correctness of timing
* Independent failures

# Characteristics of Distributed Systems

* Parallel activities - autonomous components executing concurrent tasks
* Communication via message passing - no shared memory
* Resource sharing - printer, database, other services. We need to give "fair" access to many users
* No global state - No single process can have knowledge of the current global state of the system
* No global clock - Only limited precision for processes to synchronize their clocks(can assume miliseconds accuracy at best)

    * no shared memory
    * homogeneous(identical) in hard- and software

# Distribution Transparency

To hide from the user and the application programmer the separation/distribution of components, so that the system is perceived as a whole rather than a collection of independent components.

* Access transparency - Access to local or remote resources is identical. E.g. Network File System
* Location transparency - Access without knowledge of location. DNS gives us this. If server moves and changes IP, DNS will always point to current location (with some limits)
* Failure transparency - Tasks can be completed despite failures. E.g. message retransmission, failure of a web server node should not bring down the website.
* Replication transparency - Access to replicated resources as if there was just one. And provide enhanced reliability and performance without knowledge of the replicas by users or application programmers.
* Migration(mobility/relocation) transparency - Allow the movement of resources and clients within a system without affecting the operation of users or applications. E.g. switching from one name server to another at runtime
* Concurrency transparency - A process should not notice that there are other sharing the same resources
* Performance transparency - Allows the system to be reconfigured to improve performance as loads vary. E.g. dynamic addition/deletion of components, switching from linear structures to hierarchical structures when the number of users increase
* Scaling transparency - Allows the system and application to expand in scale without changes in the system structure or the application algorithms
* Application level transparency
    * Persistence transparency - Masks the deactivation and reactivation of an object
    * Transaction transparency - Hides the coordination required to satisfy the transactional properties of operations
    
# Fault Tolerance

* Failure: an offered service no longer complies with its specification(e.g. no longer available or very slow to be usable)
* Fault: cause of a failure(e.g. crash of a component)
* Fault tolerance: no failure despites faults
